# Dirichlet Process Mixture Model for the HDM05-Sport Dataset

This script aims to perform action segmentation using DPMM trained with memoroized variational inference method. The DPMM method provides a nice way to derive model parameters and model complexity jointly. We test with the following features: **jointLocs**, **relativeAngle** and **quaternions**, since the HDM05 dataset does not have video data.

```python
import bnpy
import numpy as np
import os
import time

from matplotlib import pyplot as plt
import seaborn as sns
```

```python
def plot_segmentation(labels):
    plt.plot(labels)
    plt.show()
    
```

```python
# feature_list = ['jointLocs','relativeAngle','quaternion']
feature_list = ['relativeAngle']
dataset_path = '/mnt/hdd/Dataset_MPIIHDM05/features'
result_path = '/mnt/hdd/Dataset_MPIIHDM05/DPMM_segmentation_results'



for feature_type in feature_list:
    cpt_time = []
    for seq in range(59):
        print '-- processing: HDM05_{:s}_v{:d}'.format(feature_type, seq+1)

        ## read the features from the file
        filename = 'HDM05_{:s}_v{:d}'.format(feature_type, seq+1)
        print '--- read data'
        dataset = bnpy.data.XData.read_mat(os.path.join(dataset_path, filename))

        ## train DPMM using online algorithm
        print '--- train DPMM'
        time_start = time.time()
        K1_trained_model, K1_info_dict = bnpy.run(
        dataset, 'DPMixtureModel', 'DiagGauss', 'memoVB',
        output_path='/tmp/HDM05_DPMM/trymoves-K=1/',
        doWriteStdOut=False,
        nLap=30, nTask=1, nBatch=10,
#             gamma0=1e3, 
        sF=1e-3, ECovMat='eye',
        K=1, initname='randexamples',
        moves='birth,merge,shuffle',
        m_startLap=5, b_startLap=2, b_Kfresh=4)

        ## measure running time
        eps = time.time()-time_start
        cpt_time = np.append(cpt_time, eps)

        ## assign labels to samples (0-based cluster)
        print '--- output results'
        Q = K1_trained_model.calc_local_params(dataset)['E_log_soft_ev'] # time X dimension, matrix of encoded features
        sample_labels = np.argmax(Q, axis=1)
        plot_segmentation(sample_labels)

        ## save the sample labels to file
        outfile = 'HDM05_{:s}_v{:d}_DPMM_SampleLabels.txt'.format(feature_type, seq+1)
        np.savetxt(os.path.join(result_path, outfile), sample_labels)

        ## save the encoded features to file
        outfile2 = 'HDM05_{:s}_v{:d}_DPMM_SampleEncoding.txt'.format(feature_type, seq+1)
        np.savetxt(os.path.join(result_path, outfile2), Q)
      
    outfile3 = 'HDM05_{:s}_DPMM_Runtime.txt'.format(feature_type)
    log_runtime=np.mean(cpt_time)
    np.savetxt(os.path.join(result_path, outfile3), log_runtime.reshape(1,))
```
